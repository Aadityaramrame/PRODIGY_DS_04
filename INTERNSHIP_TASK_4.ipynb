```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
url = "https://github.com/m4gaikwad/PandasTutorial/raw/main/online_store_customer_data.csv"
data = pd.read_csv(url,on_bad_lines='skip')
data
```

```python
data.sample(5)
```

```python
data.iloc[1:5,2:6]
```

```python
data[data['State_names']== 'Kansas' ]
```

```python
data.rename(columns={"Transactiond_Date" : 'Date',
                     'State_name ': 'State'},inplace ='True')
```

```python
df = data.dropna(subset = ['Age'])
```

```python
df['Age_Range'] = np.where(df['Age'] <= 30,'Young',
                           np.where(df['Age']<= 45,'Middle',
                                    'old'))
df['Age_Range']
```

```python
df.groupby('Age_Range')['Age'].count()
```

```python
pd.crosstab(df['Age_Range'],df['Payment_method']).plot(kind='bar')

```

```python
df.Age_Range.value_counts().plot(kind='pie')
```

```python
top_likes_platform_F = data.groupby('Age')['Amount_spent'].sum().nlargest(10)
top_likes_platform_F.plot(kind='bar')
plt.title('Top Amount Spent by Each Age')
plt.xlabel('Age')
plt.ylabel('Amount_spent')
plt.show()
```

```python
pd.crosstab(df['Age_Range'],df['Payment_method']).plot(kind='box')
plt.title('Box Plot of Age_Range and Payment_method')
```

```python
# Assuming df is your DataFrame
plt.scatter(df['Age'], df['Amount_spent'])
plt.title('Scatter Plot of Age and Payment_method')
plt.xlabel('Age')
plt.ylabel('Amount_spent')
plt.show()
```
